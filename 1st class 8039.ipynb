```python
def divide(low, high):
    pivot = a[low]
    i = low + 1
    j = high
    while True:
        while i <= high and a[i] <= pivot:
            i += 1
        while j > low and a[j] > pivot:
            j -= 1
        if i < j:
            a[i], a[j] = a[j], a[i]
        else:
            break

    a[low], a[j] = a[j], a[low]
    return j
def quicksort(low, high):
    if low < high:
        p = divide(low, high)
        quicksort(low, p - 1)
        quicksort(p + 1, high)
a = [34, 23, 12, 45, 6, 7, 21, 65, 3, 2, 1, 77]
quicksort(0, len(a) - 1)
print("After sorting:")
print(a)
```

```python
def combine(low,mid,high):
    i,j=low,mid+1
    temp =[]
    while i<=mid and j<=high:
        if a[i] < a[j]:
            temp.append(a[i])
            i+=1
        else:
            temp.append(a[j])
            j+=1
    while i<=mid:
        temp.append(a[i])
        i+=1
    while j<=high:
        temp.append(a[j])
        j+=1
    i=low
    for x in temp:
        a[i] =x
        i+=1
def Mergesort(low,high):
    if low< high:
        mid=((low+high)/2)
        Mergesort(low, mid)
        Mergesort(mid+1,high)
        combine(low,mid,high)
a=[34,23,12,45,6,7,21,65,3,2,1,77]
Mergesort(0,len(a)-1)
print(f'after sorting:\n{a}')
```

```python
def display(name,dict):
    for node,value in dict.items():
        print(f'node-{node}: {value}')
from  networkx import  Graph
g=Graph()
edges=[(1,2),(1,3),(2,3),(2,4),(3,4)]
g.add_edges_from(edges)
from networkx import draw
draw(g,node_size=500,node_color='green',with_labels=True)
import matplotlib.pyplot as plt
plt.title('social network visualization')
plt.show()
from  networkx import  betweenness_centrality
display('betweenness centrality',betweenness_centrality(g))
from  networkx import  closeness_centrality
display('closeness centrality',closeness_centrality(g))
from  networkx import  degree_centrality
display('degree centrality',degree_centrality(g))
from  networkx import  pagerank
display('Pagerank',pagerank(g))
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
feature_vectors = iris.data
feature_names = iris.feature_names
species = iris.target
species_name = iris.target_names
import numpy as np
for i in range(4):
    feature = feature_vectors[:,i]
    print(f'Summary Statistics of {feature_names[i]}:')
    print(f'Mean of {feature_names[i]}: {np.mean(feature)}')
    for j in range(3):
        print(f'Mean of {feature_names[i]} of {species_name[j]}: {np.mean(feature[species==j])}')
    print(f'Max of {feature_names[i]}: {np.max(feature)}')
    print(f'Min of {feature_names[i]}: {np.min(feature)}')
    std = np.std(feature)
    print(f'Standard Deviation of {feature_names[i]}: {std}')
    print(f'Variance of {feature_names[i]}: {std*std}')
```
